```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.interpolate import pchip_interpolate
```

```python
nsteps = 100

def objective(x1, x2):
    """
    Effective average horizontal walking speed (m/s) experienced by arrivals,
    accounting for reduced free speed due to height and capacity limits due to width.
    """

    # parameters (tune these if you like)
    wmin, wmax = 2, 25 #m (minimum and maximum width of the bridg, bounds)
    hmin, hmax = 3, 10 #m (minimum and maximum height, bounds)

    q = 10.0             # arrivals (people / s)
    v0 = 1.4             # free walking speed on flat small bridge (m/s)
    j_max = 1.5          # max pedestrian flow per meter width (people / (s * m))
    beta = 0.25          # relative speed reduction across height range [hmin,hmax]; 0.25 = 25% slower at hmax

    # --- free-flow speed depending on height ---
    frac = (x2 - hmin) / (hmax - hmin)   # assume valid domain
    v_free = v0 * (1.0 - beta * frac)
    v_free = np.maximum(v_free, 0.3)  # prevent unrealistic low values
    
    # --- capacity depending on width ---
    w_dir = x1 / 2.0
    q_max = j_max * w_dir
    
    # --- effective speed (reduced if arrivals > capacity) ---
    capacity_factor = np.minimum(1.0, q_max / q)
    v_eff = v_free * capacity_factor

    return pchip_interpolate([0.16, 1, 1.4], [0, 40, 100], v_eff)

# Example: slice of width vs score at fixed height
widths = np.linspace(2, 25, nsteps)
scores = objective(widths, 3)  # keep height = 3 m

plt.plot(widths, scores)
plt.xlabel("Width (m)")
plt.ylabel("Preference score")
plt.title("Preference vs width (height = 3 m)")
plt.grid(True)
plt.show()
```
